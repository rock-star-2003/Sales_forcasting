```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('https://github.com/rock-star-2003/Datasets/raw/refs/heads/main/Project/sales_data.csv')
```

```python
df_name = pd.read_csv('https://github.com/rock-star-2003/Datasets/raw/refs/heads/main/Project/annex1.csv')
```

```python
df2 = pd.merge(df, df_name, on='Item Code', how='left')
```

```python
df2
```

```python
df = df2.iloc[:,:-2]
df
```

```python

# Replace 'Yes' with 1 and 'No' with 0 in the 'Discount (Yes/No)' column
df['Discount (Yes/No)'] = df['Discount (Yes/No)'].replace({'Yes': 1, 'No': 0})
df
```

```python
df = df.drop(columns=['Item Code'])
```

```python
# Convert the 'Date' column to datetime objects if it's not already
df['Date'] = pd.to_datetime(df['Date'])

# Group by date and item name, then sum the sales
daily_sales = df.groupby(['Date', 'Item Name'])['Quantity Sold (kilo)'].sum().reset_index()

# Print or use the daily_sales Series
daily_sales
```

```python
salesCount = df.groupby(['Date','Item Name'])['Item Name'].value_counts().rename('Sales_count').reset_index()
salesCount
```

```python
daily_sales = pd.merge(daily_sales,salesCount,on = ['Date' , 'Item Name'],how='left')
```

```python
dis = df.groupby(['Date' , 'Item Name'])['Discount (Yes/No)'].sum().reset_index()
dis
```

```python

dis['Discount (Yes/No)'] = dis['Discount (Yes/No)'].apply(lambda x: 1 if x > 0 else 0)
dis
```

```python
# Merge 'dis' with 'df' on 'Date' and 'Item Name'
daily_sales = pd.merge(daily_sales, dis, on=['Date', 'Item Name'], how='left')

# Display the merged DataFrame
daily_sales
```

```python
daily_sales['Discount (Yes/No)'].value_counts()
```

```python
price = df.groupby(['Date','Item Name','Unit Selling Price (RMB/kg)'])['Item Name'].value_counts().reset_index()
price
```

```python
# Drop duplicate rows based on 'Date' and 'Item Name', keeping the first occurrence
price = price.drop_duplicates(subset=['Date', 'Item Name'], keep='first')
price
```

```python
daily_sales = pd.merge(daily_sales,price,on=['Date','Item Name'],how='left')
```

```python
daily_sales = daily_sales.iloc[:,:-1]
daily_sales
```

```python
# Assuming 'daily_sales' DataFrame exists and has a 'Date' column
daily_sales['monthOfyear'] = pd.to_datetime(daily_sales['Date']).dt.month
daily_sales['DayOfWeek'] = daily_sales['Date'].dt.dayofweek  # Monday=0, Sunday=6
daily_sales['DayOfMonth'] = daily_sales['Date'].dt.day
daily_sales['year'] = daily_sales['Date'].dt.year
daily_sales['Weekend'] = daily_sales['DayOfWeek'].apply(lambda x: 1 if x in [5, 6] else 0)
# Calculate the week of the month
daily_sales['WeekOfMonth'] = (daily_sales['DayOfMonth'] - 1) // 7 + 1
daily_sales['quater'] = daily_sales['Date'].dt.quarter
daily_sales
```

```python

# Create an instance of LabelEncoder
encoder = LabelEncoder()

# Fit and transform the 'Item Name' column
daily_sales['Item Name'] = encoder.fit_transform(daily_sales['Item Name'])

daily_sales
```

```python
x = daily_sales.drop(columns = ['Date','Quantity Sold (kilo)','Sales_count'])
y = daily_sales.iloc[:,2]
```

```python

# Import necessary libraries (assuming they are already installed)
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression, Ridge, Lasso, ElasticNet
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor #, GradientBoostingRegressor
from sklearn.svm import SVR
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error, r2_score,mean_absolute_percentage_error

# ... (Your existing code for data preprocessing) ...

# Split data into training and testing sets (assuming x and y are defined)
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=0)
# from sklearn.preprocessing import StandardScaler
# sc = StandardScaler()
# x_train = sc.fit_transform(x_train)
# X_test = sc.transform(x_test)

# Initialize regression models
models = {
    "Linear Regression": LinearRegression(),
    "Ridge Regression": Ridge(),
    "Lasso Regression": Lasso(),
    "Elastic Net": ElasticNet(),
    "Decision Tree": DecisionTreeRegressor(),
    "Random Forest": RandomForestRegressor(),
    "Support Vector Regression": SVR(),
    "K-Nearest Neighbors": KNeighborsRegressor()
}

results = {}

for name, model in models.items():
    model.fit(x_train, y_train)  # Train the model
    y_pred = model.predict(x_test) # Make predictions

    mse = mean_squared_error(y_test, y_pred)
    r2 = r2_score(y_test, y_pred)
    mase = mean_absolute_percentage_error(y_test,y_pred)
    results[name] = {"MSE": mse, "R-squared": r2,"mase" : mase}

# Find the best model based on MSE (lower is better)
best_model_mse = min(results, key=lambda k: results[k]["MSE"])
best_model_r2 = max(results, key=lambda k: results[k]["R-squared"])
best_model_mase = min(results, key=lambda k: results[k]["mase"])


print("Model Performance:")
for name, metrics in results.items():
    print(f"{name}: MSE={metrics['MSE']:.4f}, R-squared={metrics['R-squared']:.4f}, mase={metrics['mase']:.4f}")

print(f"\nBest model based on lowest MSE: {best_model_mse}")
print(f"\nBest model based on highest R-squared: {best_model_r2}")
print(f"\nBest model based on highest best_model_mase: {best_model_mase}")
```

```python
predicted_future_sales = []
def pfs(date,data):
  weekend = 0
  if pd.to_datetime(date).day_of_week == 5 or pd.to_datetime(date).day_of_week == 6:
    weekend = 1
  # Create a DataFrame instead of a dictionary
  data_df = pd.DataFrame({
      'Item Name': [data[1]],  # Use the encoded value for 'Item Name'
      'Discount (Yes/No)': [data[2]],
      'Unit Selling Price (RMB/kg)': [data[3]],  # Example Value
      'monthOfyear': [pd.to_datetime(date).month],
      'DayOfWeek': [pd.to_datetime(date).day_of_week],
      'DayOfMonth': [pd.to_datetime(date).day],
      'year': [pd.to_datetime(date).year],
      'Weekend': [weekend],
      'WeekOfMonth': [(pd.to_datetime(date).day - 1) // 7 + 1],
      'quater': [pd.to_datetime(date).quarter]
  })

  predicted_future_sales.append([date,data[0],rf_model.predict(data_df)[0],data[2],data[3]])



from datetime import timedelta,datetime
def sales_forcasting(dt_start,dt_end,discounted):
  while dt_start <= dt_end:
    dt_start+= timedelta(days=1)
    for i in item_and_discount:
      pfs(dt_start,i)



start_date = datetime.now().date()
end_date = start_date + timedelta(days=30)

dates = sales_forcasting(start_date,end_date,0)
```
